```python
# Q1. 클로저를 사용하여 최댓값과 최솟값 찾기

def find_min_max(numbers):
    min_value = min(numbers)
    max_value = max(numbers)

    # 최솟값 반환 함수
    def find_min():
        return min_value

    # 최댓값 반환 함수
    def find_max():
        return max_value

    return find_min, find_max


# 테스트
numbers = [3, 7, 1, 9, 2, 8]

find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())
print("최댓값:", find_max())
```

```python
# Q2. 함수 호출 횟수를 출력하는 데코레이터 만들기

def counter(func):
    count = 0  # 실행 횟수 저장

    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1

        # 원래 함수 실행
        result = func(*args, **kwargs)

        # 함수 이름과 실행 횟수 출력
        print(f"{func.__name__} 실행횟수 : {count}")

        return result

    return wrapper


# 데코레이터 적용
@counter
def say_hello():
    print("Hello Aiffel!")


# 5번 실행
for i in range(5):
    say_hello()
```
