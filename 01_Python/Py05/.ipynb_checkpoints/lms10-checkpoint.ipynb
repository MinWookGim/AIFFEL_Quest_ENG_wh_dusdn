```python
class Kiosk:
    
    # 생성자
    def __init__(self):
        self.menu = ['아메리카노', '라떼', '카푸치노', '에이드']
        self.price = [2000, 3000, 3500, 4000]

        self.order_menu = []
        self.order_price = []

    # 메뉴 출력
    def menu_print(self):

        print("\n======== 메뉴판 ========")

        for i in range(len(self.menu)):
            print(i + 1, ".", self.menu[i], ":", self.price[i], "원")

        print("========================")


    # 주문
    def menu_select(self):

        while True:

            n = int(input("음료 번호 선택 (종료:0) : "))

            # 주문 종료
            if n == 0:
                print("주문을 종료합니다.")
                break

            # 메뉴 번호 검사
            elif n > len(self.menu) or n < 1:
                print("없는 메뉴입니다.")
            
            else:
                self.order_menu.append(self.menu[n - 1])
                self.order_price.append(self.price[n - 1])

                print(self.menu[n - 1], "주문 완료!")


    # 결제
    def pay(self):

        total = sum(self.order_price)

        print("\n총 금액 :", total, "원")

        money = int(input("지불 금액 입력 : "))

        if money >= total:
            print("결제 완료!")
            print("거스름돈 :", money - total, "원")

        else:
            print("금액이 부족합니다.")


    # 영수증 출력
    def table(self):

        print("\n========= 영수증 =========")

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ":", self.order_price[i], "원")

        print("--------------------------")
        print("총 합계 :", sum(self.order_price), "원")
        print("==========================")



# 객체 생성
kiosk = Kiosk()

# 실행
kiosk.menu_print()
kiosk.menu_select()
kiosk.pay()
kiosk.table()
```
