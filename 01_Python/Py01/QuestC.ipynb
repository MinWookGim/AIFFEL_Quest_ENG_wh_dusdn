```python
def is_palindrome(word):

    # 문자열을 뒤집어서 비교
    if word == word[::-1]:
        return True

    else:
        return False


# 테스트
print(is_palindrome("토마토"))   # True
print(is_palindrome("apple"))   # False
print(is_palindrome("level"))   # True
```

```python
# 회문 판독기 함수

def is_palindrome(word):

    return word == word[::-1]


# 회문 분류기 함수

def palindrome_classifier(word_list):

    # 결과 저장용 딕셔너리
    result = {
        "회문O": [],
        "회문X": []
    }

    # 회문 개수 저장 변수
    palindrome_count = 0


    # 리스트의 단어 하나씩 검사
    for word in word_list:

        # 회문이면
        if is_palindrome(word):

            result["회문O"].append(word)
            palindrome_count += 1

        # 회문이 아니면
        else:

            result["회문X"].append(word)


    # 회문 비율 계산
    ratio = palindrome_count / len(word_list)


    # 결과 반환
    return result, palindrome_count, ratio


# 테스트용 리스트
words = ["토마토", "apple", "level", "banana", "radar"]


# 함수 실행
data, count, ratio = palindrome_classifier(words)


# 결과 출력
print("분류 결과 :", data)
print("회문 개수 :", count)
print("회문 비율 :", ratio)
```

```python
# 학생 성적 데이터

scores = {

    "유원영": {
        "국어": 94,
        "영어": 91,
        "수학": 89
    },

    "설지민": {
        "국어": 88,
        "영어": 96,
        "수학": 90
    },

    "신윤아": {
        "국어": 86,
        "영어": 88,
        "수학": 77
    },

    "장윤아": {
        "국어": 98,
        "영어": 76,
        "수학": 92
    }
}


# 성적 분석 함수

def analyze_score(name):

    # 학생 정보 가져오기
    student = scores[name]


    # 과목 점수 저장
    korean = student["국어"]
    english = student["영어"]
    math = student["수학"]


    # 평균 계산
    average = (korean + english + math) / 3


    # 최고 점수 계산
    highest = max(korean, english, math)


    # 결과 출력
    print("이름 :", name)
    print("국어 :", korean)
    print("영어 :", english)
    print("수학 :", math)
    print("평균 :", average)
    print("최고 점수 :", highest)


# 함수 실행
analyze_score("유원영")
```
