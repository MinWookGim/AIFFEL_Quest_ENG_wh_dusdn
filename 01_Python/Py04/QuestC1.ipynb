```python
import json
import os


class Account:

    def __init__(self, bank, account_number, password, balance=0):

        self.bank = bank
        self.account_number = account_number
        self.password = password
        self.balance = balance

    # 입금
    def deposit(self, amount):

        if amount >= 1:

            self.balance += amount

            print(f"{amount:,}원 입금 완료")
            print(f"현재 잔액: {self.balance:,.0f}원")

        else:
            print("1원 이상 입금 가능합니다.")

    # 출금
    def withdraw(self, amount):

        if amount > self.balance:

            print("잔액이 부족합니다.")
            print(f"현재 잔액: {self.balance:,.0f}원")

        else:

            self.balance -= amount

            print(f"{amount:,}원 출금 완료")
            print(f"현재 잔액: {self.balance:,.0f}원")

    # 계좌 정보
    def display_info(self):

        print("\n===== 계좌 정보 =====")
        print(f"은행이름: {self.bank}")
        print(f"계좌번호: {self.account_number}")
        print(f"현재 잔액: {self.balance:,.0f}원")
        print("=====================\n")

    # 저장
    def save_account(self):

        accounts = load_accounts()

        key = f"{self.bank}_{self.account_number}"

        accounts[key] = {
            "bank": self.bank,
            "account_number": self.account_number,
            "password": self.password,
            "balance": self.balance
        }

        with open("accounts.json", "w", encoding="utf-8") as file:

            json.dump(accounts, file, ensure_ascii=False, indent=4)


# 전체 계좌 불러오기
def load_accounts():

    if os.path.exists("accounts.json"):

        with open("accounts.json", "r", encoding="utf-8") as file:

            return json.load(file)

    return {}


# =========================
# 로그인 / 회원가입
# =========================

while True:

    print("\n===== 은행 시스템 =====")
    print("1. 로그인")
    print("2. 회원가입")

    select = input("선택: ")

    accounts = load_accounts()

    # 로그인
    if select == "1":

        bank_name = input("은행 이름 입력: ")
        account_number = input("계좌번호 입력: ")
        password = input("비밀번호 입력: ")

        key = f"{bank_name}_{account_number}"

        # 계좌 존재 확인
        if key in accounts:

            saved_data = accounts[key]

            # 비밀번호 확인
            if password == saved_data["password"]:

                print("로그인 성공!")

                account = Account(
                    saved_data["bank"],
                    saved_data["account_number"],
                    saved_data["password"],
                    saved_data["balance"]
                )

                break

            else:

                print("비밀번호가 틀렸습니다.")

        else:

            print("존재하지 않는 계좌입니다.")

    # 회원가입
    elif select == "2":

        bank_name = input("새 은행 이름 입력: ")
        account_number = input("새 계좌번호 입력: ")
        password = input("새 비밀번호 설정: ")

        key = f"{bank_name}_{account_number}"

        # 중복 계좌 확인
        if key in accounts:

            print("이미 존재하는 계좌입니다.")

        else:

            account = Account(
                bank_name,
                account_number,
                password
            )

            account.save_account()

            print("회원가입 완료!")

            break

    else:

        print("잘못된 입력입니다.")


# =========================
# 계좌 정보 출력
# =========================

account.display_info()


# =========================
# 메뉴
# =========================

while True:

    print("===== 은행 메뉴 =====")
    print("1. 입금")
    print("2. 출금")
    print("3. 계좌 정보")
    print("4. 종료")

    menu = input("메뉴 선택: ")

    # 입금
    if menu == "1":

        amount = int(input("입금 금액 입력: "))

        account.deposit(amount)

        account.save_account()

    # 출금
    elif menu == "2":

        amount = int(input("출금 금액 입력: "))

        account.withdraw(amount)

        account.save_account()

    # 계좌 정보
    elif menu == "3":

        account.display_info()

    # 종료
    elif menu == "4":

        print("프로그램 종료")

        account.save_account()

        break

    else:

        print("잘못된 입력입니다.")
```
