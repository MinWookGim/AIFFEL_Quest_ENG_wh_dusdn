```python
import random


# =========================
# 부모 클래스
# =========================
class Character:

    def __init__(self, name, level, hp, attack, defense):

        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    # 생존 확인
    def is_alive(self):
        return self.hp > 0

    # 데미지 받기
    def take_damage(self, damage):

        real_damage = damage - self.defense

        if real_damage < 0:
            real_damage = 0

        self.hp -= real_damage

        if self.hp < 0:
            self.hp = 0

        print(f"{self.name} 이(가) {real_damage} 데미지를 받았습니다.")
        print(f"{self.name} 현재 체력: {self.hp}")

    # 공격
    def attack_target(self, target):

        # 크리티컬 확률
        critical = random.random() < 0.2

        damage = random.randint(1, self.attack)

        if critical:
            damage *= 2
            print("🔥 크리티컬 공격!")

        print(f"{self.name} -> {target.name} 공격! ({damage} 데미지)")

        target.take_damage(damage)


# =========================
# 플레이어 클래스
# =========================
class Player(Character):

    def __init__(self, name):

        super().__init__(name, 1, 120, 25, 5)

        self.exp = 0
        self.potion = 3

    # 경험치 획득
    def gain_experience(self, exp):

        self.exp += exp

        print(f"{self.name} 경험치 +{exp}")
        print(f"현재 경험치: {self.exp}")

    # 레벨업
    def level_up(self):

        while self.exp >= 50:

            self.exp -= 50

            self.level += 1
            self.attack += 10
            self.defense += 5
            self.hp += 30

            print("\n🎉 LEVEL UP!")
            print(f"레벨: {self.level}")
            print(f"공격력: {self.attack}")
            print(f"방어력: {self.defense}")
            print(f"체력: {self.hp}")

    # 포션 사용
    def use_potion(self):

        if self.potion > 0:

            heal = 40

            self.hp += heal

            self.potion -= 1

            print(f"🧪 포션 사용! 체력 +{heal}")
            print(f"남은 포션: {self.potion}")

        else:
            print("포션이 없습니다!")

    # 상태 출력
    def show_status(self):

        print("\n===== PLAYER STATUS =====")
        print(f"이름: {self.name}")
        print(f"레벨: {self.level}")
        print(f"체력: {self.hp}")
        print(f"공격력: {self.attack}")
        print(f"방어력: {self.defense}")
        print(f"경험치: {self.exp}")
        print(f"포션: {self.potion}")
        print("=========================\n")


# =========================
# 몬스터 클래스
# =========================
class Monster(Character):

    def __init__(self, name, level):

        hp = random.randint(20, 40) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(name, level, hp, attack, defense)


# =========================
# 보스 몬스터
# =========================
class BossMonster(Monster):

    def __init__(self, name, level):

        super().__init__(name, level)

        self.hp *= 2
        self.attack += 15

    # 보스 스킬
    def boss_skill(self, target):

        skill_damage = random.randint(20, 40)

        print(f"💀 {self.name} 의 보스 스킬 발동!")

        target.take_damage(skill_damage)


# =========================
# 전투 함수
# =========================
def battle(player, monster):

    print(f"\n👹 {monster.name} 등장!")
    print(f"레벨: {monster.level}")
    print(f"체력: {monster.hp}")

    turn = 1

    while player.is_alive() and monster.is_alive():

        print(f"\n===== TURN {turn} =====")

        # 플레이어 행동
        action = input("1. 공격  2. 포션 사용 : ")

        if action == "2":
            player.use_potion()

        else:
            player.attack_target(monster)

        # 몬스터 생존 시 공격
        if monster.is_alive():

            # 보스 스킬 확률
            if isinstance(monster, BossMonster):

                if random.random() < 0.3:
                    monster.boss_skill(player)

                else:
                    monster.attack_target(player)

            else:
                monster.attack_target(player)

        turn += 1

    # 승리
    if player.is_alive():

        print(f"\n✅ {monster.name} 처치 성공!")

        exp = monster.level * 20

        player.gain_experience(exp)

        player.level_up()

    else:
        print("\n❌ 전투 패배...")


# =========================
# 메인 함수
# =========================
def main():

    print("===== 자동사냥 RPG =====")

    monster_dict = {
        "슬라임": 1,
        "고블린": 2,
        "오크": 3
    }

    player_name = input("플레이어 이름 입력: ")

    player = Player(player_name)

    # 일반 몬스터 전투
    for monster_name, level in monster_dict.items():

        player.show_status()

        monster = Monster(monster_name, level)

        battle(player, monster)

        # 플레이어 사망 시 종료
        if not player.is_alive():

            print("\n💀 GAME OVER")
            return

    # 보스전
    print("\n🔥 FINAL BOSS 등장!")

    boss = BossMonster("데몬킹", 5)

    battle(player, boss)

    if player.is_alive():
        print("\n🏆 게임 클리어!")
    else:
        print("\n💀 GAME OVER")


# =========================
# 게임 시작
# =========================
main()
```
